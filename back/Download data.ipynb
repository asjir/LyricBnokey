```python
artists = [ 'slipknot','prodigy','system of a down','marina','selena gomez','modern talking','imagine dragons','metallica','britney spears','kiss','bhad barbie','my chemical romance','rammstein','pink floyd','joy division','bones','our last night','twenty one pilots','conan gray','drake','geazy','joji','billie eilish','david bowie','ariana grande','travis scott','jack stauber','sting','paramore','red hot chilli peppers','the weeknd','lil peep','avicii','xxxtentacion','adele','niall horan','loic nottet','disturbed','bring me the horizon','queen','ghostemane','chase atlantic','onerepublic','die antwood','depeche mode','hollywood undead','frank ocean','asap rocky','blackbear','logic','oasis','liam gallagher',' halsey','aurora','tom odell','arctic monkeys','the neighbourhood','daft punk','mac demarco','tame impala','radiohead','lana del rey','yungblud','troye sivan','alice in chains','pristin','kreayshawn','u2','banks','eminem','beyonce','blackpink','schafter']
def artist_list_to_track_ids(alist):
    track_ids = []
    for a in alist:
        tracks = requests.get(url+'track.search', {'apikey':apikey, 'q_artist':a.encode('utf-8')}).json()['message']['body']['track_list'][:20]
        for track in tracks:
            if track['track']['has_lyrics']==1:
                track_ids += [track['track']['track_id']]
            else:
                continue
    return(track_ids)
```

```python
print(len(artists))
trids=[]
trids = artist_list_to_track_ids(artists) #trids reset to [] to make sure I won't waste any calls now
np.save('trids', trids)
```

```python
allyrics=[]
for trid in trids:
    lyrics = requests.get(url+'track.lyrics.get', {'apikey':apikey, 'track_id':trid})
    allyrics+=[(lyrics.json()['message']['body']['lyrics']['lyrics_body'], trid)]
```

```python
np.save('allyrics', np.array(allyrics))
```
